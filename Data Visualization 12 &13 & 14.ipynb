#### box and whisker plot

- numerical variable
- five-number summary
   a)min
   b)First quartile (Q1)
   c)Second quartile (Q2) - median
   d)Third quartile (Q3)
   e)max

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/gchoi/Dataset/master/ToyotaCorolla.csv') 
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

#### case 1: numerical variable

```python
sns.boxplot(data=df,y='Price')
```

#### case 2 : Numerical VS Categorical 

```python
sns.boxplot(data=df,x='FuelType',y='Price')
```

#### case 3 : FuelType vs Price and Automatic

```python
sns.boxplot(data=df,x='FuelType',y='Price',hue='Automatic')
```

#### case 4: Box-whisker plot and Histogram

```python
f,(ax_box,ax_hist)=plt.subplots(2,gridspec_kw={'height_ratios':{.15,.85}})
sns.boxplot(data=df,x='Price',ax=ax_box,)
sns.histplot(data=df,x='Price',ax=ax_hist,kde=True)
plt.show()
```

#### Pairwise Plots

- It is used to plot pairwise relationships in a dataset.

- creates scatterplot for joint relationships
- creates histograms for univariate distributions

```python
sns.pairplot(data=df,kind='scatter',hue='FuelType')
```
