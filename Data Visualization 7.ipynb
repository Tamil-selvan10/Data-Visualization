### Barplot

- Barplot   ->  frequency distribution of categorical variable
- Histogram ->  frequency distribution of numerical variable


- A Bar plot is a plot that represents categorical data with rectangular bars 
with length proprotions to the counts that they represent.

### Case 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=[23,45,56,78,213]
plt.bar([1,2,3,4,5],data)
plt.grid()
plt.show()
```

#### case 2 - color

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=[23,45,56,78,213]
plt.bar([1,2,3,4,5],data,color=['red','blue','cyan','green','yellow'])
plt.show()
```

#### case 3- range

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=[10,20,30,40,50]
plt.bar(range(len(data)),data,color='green')
plt.show()
```

#### Case 4 - alpha attribute

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=[10,20,30,40,50]

plt.bar(range(len(data)),data,color='royalblue',alpha=0.5)
plt.grid(color='red',linestyle='--',linewidth=2,axis='y',alpha=0.5)
plt.show()
```

#### case 5 - xticks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=[10,20,30,40,50]
labels='A B C D E'.split()

plt.title('Title of the bar plot')
plt.bar(range(len(data)),data)
plt.xticks(range(len(data)),labels)
plt.xlabel('category')
plt.ylabel('cost')
plt.show()
```

#### case 6 - width

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data1=[30,59,60,49,56]
data2=[49,53,71,66,59]
width=0.3

plt.bar(np.arange(len(data1)),data1,width=width)#data1
plt.bar(np.arange(len(data2))+width,data2,width=width)#data2
plt.grid(color='red',linestyle='--')
plt.show()
```
