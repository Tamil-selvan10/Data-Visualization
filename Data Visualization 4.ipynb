#### Case 1

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,10,100)

#Figure 1
plt.figure(1)

plt.subplot(2,2,1)
plt.title('Linear')
plt.plot(x,x,'b--')

plt.subplot(2,2,2)
plt.title('sin(x)')
plt.plot(x,np.sin(x),'r--')


plt.subplot(2,2,3)
plt.title('cos(x)')
plt.plot(x,np.cos(x),'y--')

plt.subplot(2,2,4)
plt.title('exp(x)')
plt.plot(x,np.exp(x),'k--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=np.sin(x)

# plt.plot(x,y,lw=2)
plt.plot(x,y,linewidth=8)
plt.grid()
plt.show()
```

#### Case 3

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2*np.pi)
y1=np.sin(x)
y2=np.cos(x)


fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y1,c='b',label='y1',linewidth=3)
ax.plot(x,y2,c='r',label='y2',linewidth=3)

plt.grid()

plt.legend()

plt.show()
```

#### Case 4

```python
import matplotlib.pyplot as plt

x1=[1,2,3,4]
y1=[pow(i,3) for i in x1]

x2=[1,2,3,4]
y2=[i*3 for i in x2]

lines=plt.plot(x1,y1,x2,y2)

plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],color='b',linewidth=3.0)

plt.grid()

plt.show()
```
