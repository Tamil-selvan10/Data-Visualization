#### case 1

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,10,100)
y=np.sin(x)

#Figure 1 
plt.figure(1) #create a new figure
plt.subplot(1,2,1) #subplot(rows,columns,x)
plt.title('y=sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)

plt.subplot(1,2,2)
plt.title('y=sin(x)**2')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.plot(x,y**2)

plt.show()

```

#### case 3

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,10,100)

# Figure 1
plt.figure(1)
plt.subplot(2,2,1) #subplot(rows,columns,x)
plt.title('linear')
plt.plot(x,x,'b--')

plt.subplot(2,2,4)
plt.title('sin(x)')
plt.plot(x,np.sin(x),'rs')


# Figure 2
plt.figure(2)
plt.subplot(2,2,1)
plt.title('cos(x)')
plt.plot(x,np.cos(x),'y--')


plt.subplot(2,2,4)
plt.title('Exponential(x)')
plt.plot(x,np.exp(x),'k--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,10,100)

#Figure 1
plt.figure(1)
plt.subplot(1,2,1)
plt.title('linear')
plt.plot(x,x,'b--')


plt.subplot(1,2,2)
plt.title('sin(x)')
plt.plot(x,np.sin(x),'r--')


#Figure 2
plt.figure(2)
plt.subplot(1,2,1)
plt.title('cos(x)')
plt.plot(x,np.cos(x),'y--')

plt.subplot(1,2,2)
plt.title('exp(x)')
plt.plot(x,np.exp(x),'b--')

plt.show()
```
