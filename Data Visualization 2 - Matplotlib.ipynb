```python
plt.title('my plot on cubes')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.plot([1,2,3,4,5],[1,8,27,64,125])
plt.grid()
plt.show()
```

#### case 3: linspace

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(5,100,10)
print(x)
y=np.linspace(10,1000,10)
print(y)

plt.plot(x,y)
plt.show()

#Note: x & y length should be same.
```

#### case4: x and y limits

```python
# Specifying the limits of X and y labels as a range 
# xlim[xmin,xmax]
# ylim[ymin,ymax]
```

```python
import matplotlib.pyplot as plt
data1=[11,12,13,14,15,16,17]
data2=[15.5,12.5,11.7,9.50,12.50,11.50,14.75]

# Add Title
plt.title('Interactive Plot')

# Add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# set limit for each axis
plt.xlim(11,17)
plt.ylim(9,16)

plt.plot(data1,data2)

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0.,5.,0.2)
print(x)

# blue dashes, red squares and green triangles
plt.plot(x,x**3,'b--',label='^3')
plt.plot(x,x**3.2,'rs',label='^3.2')
plt.plot(x,x**3.5,'g^',label='^3.5')

plt.legend(loc='best')
plt.grid()
```
