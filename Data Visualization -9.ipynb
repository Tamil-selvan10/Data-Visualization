```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/gchoi/Dataset/master/ToyotaCorolla.csv')
```

```python
import seaborn as sns
```

#### case 1

```python
sns.set(style='darkgrid') # theme of the background
sns.regplot(data=df,x='Age',y='Price') #regression plot
```

#### case 2  - No regression line

```python
sns.set(style='darkgrid')
sns.regplot(data=df,x='Age',y='Price',fit_reg=False) # by default, fit_reg=True
```

#### case 3 - marker

```python
sns.set(style='darkgrid')
sns.regplot(data=df,x='Age',y='Price',marker='*',fit_reg=False)
```

#### case 4 - age VS price by FuelType

```python
sns.set(style='darkgrid')
sns.lmplot(data=df,x='Age',y='Price',fit_reg=False,hue='FuelType',legend=True,palette='Set1')
```
