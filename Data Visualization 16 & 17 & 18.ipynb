#### Violinplot (boxplot +KDE)

#### Seaborn figure styles

- darkgrid (by default)
- whitegrid
- dark
- white
- ticks.

#### case 1 - whitegrid

```python
import seaborn as sns
sns.set(style='whitegrid')
sns.violinplot(data=df,y='Age')
```

#### case 2 - dark

```python
import seaborn as sns
sns.set(style='ticks') # style=darkgrid,whitegrid,dark,white,ticks
sns.violinplot(data=df,x='Age')
```

#### case 3 - Draw a vertical violinplot grouped by a categorical variable

```python
import seaborn as sns
sns.violinplot(data=df,x='FuelType',y='Age')
```

```python
import seaborn as sns
sns.palplot(sns.color_palette('muted'))  # 1.muted, 2.bright, 3.pastel, 4.dark, 5.colorblind, 6.deep
```

#### case 4

```python
sns.violinplot(data=df,x='FuelType',y='Age',hue='Automatic',palette='muted')
```

#### case 5 - split

```python
sns.violinplot(data=df,x='FuelType',y='Age',hue='Automatic',palette='muted',split=True) # by default, split=False
```

#### case 6 - order

```python
sns.violinplot(data=df,x='FuelType',y='Age',hue='Automatic',split=True,order=['CNG','Diesel','Petrol'])
```
